```python
d    = [1,3,5,7,10,13,15,18,20]
dmypc = [15,10,8,7,6,6,5,5,5]
dpc3  = [16,10,8,7,6,6,6,5,5]
```

```python
def theorical(N,k):
    d = []
    for i in range(len(N)):
        #d.append( math.log(k, N[i])/2 )
        d.append(k**(1/(2*N[i])))
        
    return d
```

```python
x = np.arange(2,15,.1)
mytheod = theorical(x,128*10e9)
pc3theod = theorical(x,128*2*10e9)
```

```python
plt.plot(x,mytheod,'-',color='crimson',label='Theorical maximum allocate')
plt.plot(N,myd-1,'x',label='Maximum allocation')
plt.xlabel('N')
plt.ylabel('d')
plt.yscale('log')
plt.title('Limits of allocate')
plt.legend()
```
