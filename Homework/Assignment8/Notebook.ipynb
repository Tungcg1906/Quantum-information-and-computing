Consider 𝑁 spin-1/2 particles on a one-dimensional lattice, described by the Hamiltonian
        

$$ \hat H = \lambda \sum_{i}^{N} \sigma^{z}_i +  \sum_{i}^{N-1} \sigma^{x}_i \sigma^{x}_{i+1} $$
            
where $\sigma_x$ and $\sigma_z$ are the Pauli matrices and $\lambda$ is the interaction strength

(a) Compute the ground state energy as a function of the transverse field 𝜆 by means of the real-space
RG algorithm.

(b) Optional: Compute the ground state energy as a function of 𝜆 by means of the INFINITE DMRG
algorithm. Compare the results between them and with the mean field solution.
